```python
# --- 1. Get the Number of Subjects ---
print("Welcome to the Simple GPA Calculator!")
print("This will help you calculate your GPA for the semester.")

# We use a loop to make sure the user enters a valid number
while True:
    try:
        # Get the total number of subjects the student is taking
        num_subjects = int(input("Enter the total **number of subjects** this semester: "))
        if num_subjects > 0:
            break
        else:
            print("Please enter a number greater than 0.")
    except ValueError:
        print("Invalid input. Please enter a whole number.")

# --- 2. Initialize Variables for Calculation ---
# Total grade points earned (e.g., 8*4 + 9*3, etc.)
total_grade_points = 0.0

# Total credits the student is enrolled in
total_credits_enrolled = 0

# --- 3. Get Marks and Credits for Each Subject ---
print("\n--- Now, let's enter the details for each subject ---")

# We use a 'for loop' to repeat the process for every subject
for i in range(1, num_subjects + 1):
    print(f"\n**Subject {i}:**")

    # Get the marks
    while True:
        try:
            # We use float here in case marks have decimal points
            marks = float(input(f"Enter marks (out of 100) for Subject {i}: "))
            if 0 <= marks <= 100:
                break
            else:
                print("Marks must be between 0 and 100.")
        except ValueError:
            print("Invalid input. Please enter a number for marks.")

    # Get the credits
    while True:
        try:
            # Credits are usually whole numbers
            credits = int(input(f"Enter credit value for Subject {i}: "))
            if credits >= 0:
                break
            else:
                print("Credits cannot be negative.")
        except ValueError:
            print("Invalid input. Please enter a whole number for credits.")

    # --- 4. Determine Grade Point based on Marks (Simple B.Tech Scheme) ---
    # This is a simple 'if-elif-else' chain to map marks to a grade point (GP)
    grade_point = 0
    if marks >= 90:
        grade_point = 10  # O (Outstanding)
        print("Grade: O (Outstanding) - Grade Point: 10")
    elif marks >= 80:
        grade_point = 9   # A+ (Excellent)
        print("Grade: A+ (Excellent) - Grade Point: 9")
    elif marks >= 70:
        grade_point = 8   # A (Very Good)
        print("Grade: A (Very Good) - Grade Point: 8")
    elif marks >= 60:
        grade_point = 7   # B+ (Good)
        print("Grade: B+ (Good) - Grade Point: 7")
    elif marks >= 50:
        grade_point = 6   # B (Above Average)
        print("Grade: B (Above Average) - Grade Point: 6")
    elif marks >= 45:
        grade_point = 5   # C (Average)
        print("Grade: C (Average) - Grade Point: 5")
    elif marks >= 40:
        grade_point = 4   # P (Pass) - **Usually minimum for passing**
        print("Grade: P (Pass) - Grade Point: 4")
    else:
        grade_point = 0   # F (Fail)
        print("Grade: F (Fail) - Grade Point: 0")


    # --- 5. Accumulate Totals ---
    # Total Points = (Grade Point * Subject Credits)
    subject_points = grade_point * credits
    total_grade_points += subject_points
    total_credits_enrolled += credits

# --- 6. Calculate Final GPA and Display Result ---
print("\n" + "="*40)
print("              **RESULTS**")
print("="*40)

if total_credits_enrolled > 0:
    # Formula for GPA: Sum(Grade Point * Credits) / Sum(Credits)
    gpa = total_grade_points / total_credits_enrolled
    # Use f-string formatting to show the GPA with 2 decimal places
    print(f"Total Grade Points Earned: **{total_grade_points:.2f}**")
    print(f"Total Credits Enrolled: **{total_credits_enrolled}**")
    print(f"\nYour Semester GPA is: **{gpa:.2f}**")

    # Final message based on GPA
    if gpa >= 9.0:
        print("üéâ Outstanding! Keep up the excellent work.")
    elif gpa >= 7.0:
        print("üëç Very Good! You're doing great.")
    else:
        print("Keep pushing! Aiming for better next semester.")
else:
    print("Error: Could not calculate GPA as no credits were entered.")

print("="*40)
```
